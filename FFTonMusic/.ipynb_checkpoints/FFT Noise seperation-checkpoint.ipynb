# NOISE REMOVAL FROM MUSIC USING FAST FOURIER TRANSFORM (FFT)

## We start by importing the necessary libraries :

### 1. Librosa - For audio processing and spectral analysis.
### 2. Numpy - For performing Fast Fourier Transform FFT, manipulate arrays, and handle numerical computations related to audio processing.
### 3. Scipy.io - To read and write WAV audio files and for loading the input audio file and saving the modified or filtered audio.
### 4. Matplotlib.pyplot - For data visualization i.e. Magnitude vs Frequency Plotting
### 5. Scipy.signal & Windows - For signal processing. Windows are often applied to data before performing operations like filtering or Fourier analysis for reducing artifacts in the audio signal during modifications.

```python
import librosa
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt
from scipy.signal import windows
```

### Now we define a function to obtain the frequency plots of various audio files to have a better understanding about how FFT works in removing noise from the audio.

```python
def freqplot(audio,name):  
    # Load the audio file
    audio_path = audio
    y, sr = librosa.load(audio_path, sr=None)

    # Compute the FFT
    fft = np.fft.fft(y)
    freqs = np.fft.fftfreq(len(fft), 1/sr)

    # Plot the magnitude spectrum
    plt.figure(figsize=(10, 5),dpi=720)
    plt.plot(freqs, np.abs(fft))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    #plt.title('Magnitude Spectrum of Audio Signal : ')
    plt.title(name)
    plt.xlim(0, sr/2)  # Display up to the Nyquist frequency
    plt.show()
```

### First we plot the the frequency spectrum of the audio without the noise and then we plot the spectrum of the file with noise added to it.

```python
freqplot('Pureaudio.wav','Pureaudio Frequency Spectrum')
freqplot('Noiseaudio.wav','Noiseaudio Frequency Spectrum')
```

### As we look into the plots, we understand that there is a noise attributed to the partcular frequency range 9-11 KHz with a peak at 10KHz and a very high magnitude. Hence we will try to remove it through FFT. The function 'fourierfilter' does just that! Note that we can specify the particular range of target frequency within the function itself!

```python
def fourierfilter(audio):
    # Load the audio file
    audio_path = audio
    y, sr = librosa.load(audio_path, sr=None)

    # Define frequency range to remove
    start_freq = 9000  # Hz
    end_freq = 11000   # Hz

    # Compute the FFT
    fft = np.fft.fft(y)
    freqs = np.fft.fftfreq(len(fft), 1/sr)

    # Find the indices corresponding to the frequencies within the range to remove
    freq_range_idx = np.where((freqs >= start_freq) & (freqs <= end_freq))[0]

    # Apply a windowed filter to reduce artifacts
    window = windows.hamming(len(freq_range_idx))
    fft[freq_range_idx] = fft[freq_range_idx] * (1 - window)

    # Inverse FFT to get the filtered signal
    filtered_signal = np.fft.ifft(fft)

    # Save the filtered signal to a new audio file
    filtered_audio_path = 'First_peak_removed.wav'
    wavfile.write(filtered_audio_path, sr, filtered_signal.real)
fourierfilter('Noiseaudio.wav')
```

### To get substantial reduction in magnitude, it is always best to repeat the process a couple of times. Hence we employ a for loop.

```python
for i in range(100):
    fourierfilter('First_peak_removed.wav')
```

### To see whether the frequency peak of 10KHz is removed, we employ the function 'freqplot'

```python
freqplot('First_peak_removed.wav','Frequency Spectrum After removing First Peak')
```

### As you can see, the first noise peak corresponding to the frequency range of 10 KHz has been successfully removed! But we also identify a second noise peak corresponding to the frequency range of 17-18KHz in the file. To remove that again we need to seek the help of 'fourierfilter' function.

```python
def fourierfilter(audio):
    # Load the audio file
    audio_path = audio
    y, sr = librosa.load(audio_path, sr=None)

    # Define frequency range to remove
    start_freq = 17800  # Hz
    end_freq = 25000   # Hz

    # Compute the FFT
    fft = np.fft.fft(y)
    freqs = np.fft.fftfreq(len(fft), 1/sr)

    # Find the indices corresponding to the frequencies within the range to remove
    freq_range_idx = np.where((freqs >= start_freq) & (freqs <= end_freq))[0]

    # Apply a windowed filter to reduce artifacts
    window = windows.hamming(len(freq_range_idx))
    fft[freq_range_idx] = fft[freq_range_idx] * (1 - window)

    # Inverse FFT to get the filtered signal
    filtered_signal = np.fft.ifft(fft)

    # Save the filtered signal to a new audio file
    filtered_audio_path = 'Second_peak_removed.wav'
    wavfile.write(filtered_audio_path, sr, filtered_signal.real)
fourierfilter('First_peak_removed.wav')
```

```python
for i in range(100):
    fourierfilter('Second_peak_removed.wav')
```

### To see whether the second noise peak has been removed, we use 'freqplot' function

```python
freqplot('Second_peak_removed.wav','Frequency Spectrum After removing Second Peak')
```

### We see that the second noise peak has also been substantially reduced and we get a file much closer in resemblance to the original file!

### A closer look though, reveals that the removal of the 2 peaks was not perfect and there is still some noise spread through a rangee of frequencies. But it is barely noticeable in the final audio file.

## This is rather a crude way of visualizing the use of FFT on filtering audio files. In actual practice, musical files will have much more overlapping audio frequencies and it would be difficult to hand pick the frequencies related to noise from such a complex dataset. Hence the results rarely meet our expectations. The file that I used here is of a 30 second audio clip into which I added noise profiles corresponding to 9-11KHz and 15-20+KHz frequencies. Hence it was easier to do FFT to visualize, identify and remove the said audio frequencies from the output. 

## Other techniques like Adaptive Filtering, Non-Negative Matrix Factorization (NMF), Deep Learning-Based Source Separation, Spectral substraction, etc can be used to have much more improved results. Best results are often obtained via combining multiple methods!
